```python

import pandas as pd


```

```python
# Load the dataset
df=pd.read_csv(r'C:\Users\Admin\Desktop\project\customer_shopping_behavior.csv') 
```

```python
#to get top 5 rows of the data
df.head()
```

```python
# to get the summary of the data
df.describe()
```

```python
#to get null values in the data
df.isnull().sum()
```

```python
#to replace null values(review rating ) using median
#review rating affects each category differently so we will replace null values with median of each category

df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum().sum() # to get total null values in the data
```

```python
#rename column names
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
#to get column names
df.columns
```

```python
#create colum age_group to categorize customers into age groups
labels=['young_adult','adult','middle-aged','senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)  #young adult: 18-30, adult: 31-45, middle-aged: 46-60, senior: 61 and above   
```

```python
#want to see how q is categorized
# print q values
#young adult: 18-30, adult: 31-45, middle-aged: 46-60, senior: 61 and above   
print(pd.qcut(df['age'],q=4).unique())

```

```python
#to see age and age group columns
df[['age','age_group']].head(10)
```

```python
#to categorize customers based on frequency of purchases
df['frequency_of_purchases'].unique()
```

```python
#create purchse_frequency_day column to categorize customers into purchase frequency groups
frequency_mapping={
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Annually': 365,
    'Every 3 Months': 90,
    'Bi-Weekly': 14
}

df['purchase_frequency_day']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
#to see purchase frequency day and frequency of purchases columns
df[['purchase_frequency_day','frequency_of_purchases']].head(10)
```

```python
#to drop duplicates in purchase frequency day and frequency of purchases columns and see unique values
df[['purchase_frequency_day','frequency_of_purchases']].drop_duplicates().head(10)
```

```python
#to see discount applied and promo code used columns
df[['discount_applied','promo_code_used']].head(10)
```

```python

(df['discount_applied'] == df['promo_code_used']).all() #check if discount_applied and promo_code_used are the same
```

```python

# Dropping promo code used column : since discount_applied and promo_code_used are the same we can drop one of them to avoid redundancy

df = df.drop('promo_code_used', axis=1)
```
